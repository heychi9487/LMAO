```python
import matplotlib.pyplot as plt
student1={'chinese':78,
          'english':76,
          'math':58,
          'science':46,}
subjects=student1.keys()
score1=student1.values()

figure = plt.figure(figsize=(8,5))
print(type(figure))
axes = figure.add_subplot(1,1,1)
axes.plot(subjects,score1,'ro--')

plt.show()
```

```python
import matplotlib.pyplot as plti
from matplotlib.font_manager import FontProperties
font=FontProperties(fname=r'c:\windows\fonts\simsun.ttc',size=8)




import matplotlib.pyplot as plt

student1 = {'':78, 
            'english':80,
            'math':92,
            'history':75,
            'discover':85}

student2 = {'':69, 
            'english':92,
            'math':74,
            'history':69,
            'discover':93}

subjects = list(student1.keys())
score1 = list(student1.values())

figure = plt.figure(figsize=(8,5))   #figure function Figure(class)
axes = figure.add_subplot(1,1,1)     #figure function  add_subplot(method)Axes(class)
print(type(axes))
axes.plot(subjects,score1,'ro--',label='1')

score2 = list(student2.values())   
axes.plot(subjects,score2,'bo-',label='2')   

axes.set_title('',fontproperties=font)
axes.set_xlabel('',fontproperties=font)
axes.set_ylabel('',fontproperties=font)
axes.set_xticklabels(subjects,fontproperties=font)
axes.grid(axis='y')
axes.legend(prop=font)                

x_locations = axes.get_xticks()       #ticks
print(x_locations)
for i in x_locations:
    axes.annotate(str(score1[i]),xy=(i,score1[i]+0.5))
    axes.annotate(str(score2[i]),xy=(i,score2[i]+0.5))

plt.show()
```
